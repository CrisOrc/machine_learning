## Función

````python
def plot_predict_model(data_x, data_y, ax=None, model='linear', kernel='SVC', offset=0.5, n_neighbors = 7 ):

  import numpy as np
  kernels = ['linear', 'poly', 'rbf', 'sigmoid']

  h_min, h_max = min(data_x[:, 0]) - offset, max(data_x[:, 0]) + offset
  v_min, v_max = min(data_x[:, 1]) - offset, max(data_x[:, 1]) + offset

  h_grid, v_grid = np.meshgrid(np.linspace(h_min, h_max, 100), np.linspace(v_min, v_max, 100))

  if ax is None:
      _, ax = plt.subplots(figsize=(5, 4))

#Conditional predict model
  #logistic regresion
  if model == 'logisticRegression':
    #---Imports---#
    from sklearn.linear_model import LogisticRegression
    clf = LogisticRegression()

  #Type model Naive Bayes
  if model == 'naive_bayes':
    #---Imports---#
    from sklearn.naive_bayes import GaussianNB
    clf=GaussianNB()

  #Type model KNN
  if model == 'knn':
    #---Imports---#
    from sklearn.neighbors import KNeighborsClassifier as kNN
    clf = kNN(n_neighbors = n_neighbors , metric = "euclidean")

  #Type model SVC
  if model in kernels and kernel == 'SVC':
    #---Imports---#
    from sklearn.svm import SVC
    clf = SVC(kernel = model , gamma = 1)

  #Type model SVR
  if model in kernels and kernel == 'SVR' :
    #---Imports---#
    from sklearn.svm import SVR
    clf = SVR(kernel = model)

  #Graphic background color
  clf.fit(data_x, data_y)
  pred_grid = clf.predict(np.c_[h_grid.ravel(), v_grid.ravel()])
  pred_grid = pred_grid.reshape(h_grid.shape)
  ax.pcolormesh(h_grid, v_grid, pred_grid, cmap='Paired')
  ax.scatter(data_x[:, 0], data_x[:, 1], c=data_y, edgecolors='k', cmap='Paired')
````

## In Synthetic data

````python
import numpy as np # Librería mateamtica
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import make_regression
````

````python
X, y = make_regression(n_samples = 300, n_features = 1, noise = 20, random_state = 1)

_, ax = plt.subplots (figsize=(6,5))
ax.scatter( X , y , c = 'b', edgecolors='k')

print(X.shape, y.shape)
````

````python
plot_predict_model(data_x=X, data_y=y, model='linear', offset=0.1)
plt.tight_layout()
plt.show()
````

````python
svr = SVR(kernel = 'linear' )

svr.fit( X.reshape(-1, 1) , y )
````

````python
# Predecir la nueva muestra. Imprimir la prediccion
prediction = svr.predict(np.array([-2]).reshape(-1, 1))
print (prediction.shape)
````

````python
# Predict x values
predictions = svr.predict(X)

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , color = 'blue')
ax.plot ( X, predictions, color = 'red')
````

#Non-Linear Data

````python
import numpy as np # Librería mateamtica
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import make_regression
````

````python
X, y = make_regression(n_samples = 300, n_features = 1, noise = 20, random_state = 1)

y = (y**2) * 0.001

print(X.shape, y.shape)
_, ax = plt.subplots (figsize=(6,5))
ax.scatter( X , y , c = 'b', edgecolors='k')
````

````python
svr = SVR(kernel = 'rbf', C=100 )

svr.fit( X , y )
````

````python
predictions = svr.predict(X)

_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , color = 'blue', edgecolors='k')
ax.scatter( X, predictions, c = 'red', s=5)
````
